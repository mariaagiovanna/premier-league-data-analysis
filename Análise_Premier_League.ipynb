# Análise da Premier League 2018-2019

```python
import pandas as pd
df = pd.read_csv("E0.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
#Gols feitos em casa
home_goals = df.groupby("HomeTeam")["FTHG"].sum()
#Gols feitos fora
away_goals = df.groupby("AwayTeam")["FTAG"].sum()
#Somar os dois
total_goals = home_goals + away_goals
#Ordenar
total_goals = total_goals.sort_values(ascending=False)
total_goals.head(10)
```

```python
import matplotlib.pyplot as plt

plt.figure()
total_goals.head(10).plot(kind="bar")

plt.title("Top 20 times com mais gols - Premier League 2018/19")
plt.xlabel("Times")
plt.ylabel("Total de Gols")

plt.xticks(rotation=45)
plt.show()
```

```python
#Total de gols na temporada
total_goals_temporada = df["FTHG"].sum() + df["FTAG"].sum()
#Total de jogos
total_jogos = len(df)
#Média de gols por jogo
media_gols = total_goals_temporada / total_jogos
media_gols

#Resultados gerados --> A temporada de 2018/19 apresentou média de 2.68 gols por partida,
# indicando um campeonato ofensivamente ativo e com alto volume de oportunidades.
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.hist(df["FTHG"]+ df["FTAG"])

plt.title("Distribuição de Gols por Partida - Premier League 2018/2019")
plt.xlabel("Número de Gols no Jogo")
plt.ylabel("Quantidade de Jogos")

plt.show()
```
